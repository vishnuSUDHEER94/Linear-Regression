```
#Importing the pandas library for reading our dataset
import pandas as pd
#Getting the train test split packages from sklearn and preparing the dataset to train and test the model
from sklearn.model_selection import train_test_split
#Importing the numpy library for manipulating and working with the data
import numpy as np
#Importing the matplotlib library for plotting the graph
import matplotlib.pyplot as plt
```

```
#It reads the csv file from the given path
dataset = pd.read_csv('/content/housing.csv')
#It removes any entries (or) fields which are incompleted
dataset = dataset.dropna()
#Prints the first 10 rows of the dataset
print("Here are the first ten rows of the dataset:")
dataset.head(10)
```

```
#It plots using the dataset
dataset.plot(subplots=True)
#It adjusts subplot params so that the subplots fits into the area 
plt.tight_layout()
#Displays the figure
plt.show()
```

```
#We predicts "median_house_value" column
Y = dataset['median_house_value']
#Selects from "longitude" column to the "median_income" column
X = dataset.loc[:,'longitude':'median_income']
```

```
#Splits the data, where 70% is used for training the model and 30% is used in testing the model
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
#It Converts the datasets to numpy arrays where it can work with the PyTorch model
x_train_np = x_train.to_numpy()
y_train_np = y_train.to_numpy()
#Converting the testing data
x_test_np = x_test.to_numpy()
y_test_np = y_test.to_numpy()
```

```
#Importing the PyTorch library
import torch
#Importing the 1D Convolutional layer 
from torch.nn import Conv1d
#Importing the Max pooling layer
from torch.nn import MaxPool1d
#Importing the Flatten layer
from torch.nn import Flatten
#Importing the Linear layer
from torch.nn import Linear
#Importing the relu activation function
from torch.nn.functional import relu
#Importing the DataLoader and TensorDataset libraries from PyTorch to work with our datasets 
from torch.utils.data import DataLoader, TensorDataset
```

```
#The Class should be the subclass of torch.nn.Module
class CnnRegressor(torch.nn.Module):
  #Defining the initalization method
  def __init__(self, batch_size, inputs, outputs):
    #Initializing the superclass and storing the parameters there
    super(CnnRegressor, self).__init__()
    #Defining the batch_size
    self.batch_size = batch_size
    #Defining the inputs
    self.inputs = inputs
    #Defining the outputs
    self.outputs =outputs
    #Defining the input layer that has input channels, output channels and kernel size
    self.input_layer = Conv1d(inputs, batch_size, 1)
    #Defining the max pooling layer that has kernel size
    self.max_pooling_layer = MaxPool1d(1)
    #Defining the first conv layer     
    self.conv_layer = Conv1d(batch_size, 128, 1)
    #Defining the second conv layer
    self.conv_layer2 = Conv1d(128, 64, 1)
    #Defining the flatten layer
    self.flatten_layer = Flatten()
    #Defining the liner layer
    self.liner_layer = Linear(64, 32)
    #Defining the output layer
    self.output_layer = Linear(32, outputs)
  #Defining a method feed which is used to feed the inputs to the model
  def feed(self, input):
    #We are reshaping the entry so that it can feed to the input layer
    input = input.reshape((self.batch_size, self.inputs, 1)) 
    #It gets the output from the first layer and run it through the relu activation function    
    output = relu(self.input_layer(input))
    #It gets the output from the max pooling layer
    output = self.max_pooling_layer(output)
    #It gets the output from the first convolution layer and runs it through the relu activation function
    output = relu(self.conv_layer(output))
    #It gets the output from the second convolution layer and runs it through the relu activation function
    output = relu(self.conv_layer2(output))
    #It gets the output from the flatten layer
    output = self.flatten_layer(output)
    #It gets the output from the linear layer and runs it through the relu activation function
    output = self.liner_layer(output)
    #It gets the output from the output layer 
    output = self.output_layer(output)
    #It returns the output
    return output
```

```
#Import SGD package from PyTorch for our optimizer	
from torch.optim import SGD
#Import L1Loss package from PyTorch for our performance measure
from torch.nn import L1Loss
#Import the R^2Score package
!pip install pytorch-ignite
#Import the R^2Score package from PyTorch's ignite for the score measure
from ignite.contrib.metrics.regression.r2_score import R2Score
```

```
#It Defines the batch_size
batch_size = 128
#It has the parameters like batch_size, X columns and Y columns
model = CnnRegressor(batch_size, X.shape[1], 1)
#It sets the model which uses GPU for processing
model.cuda()
```

```
#This method will retun retun the L1Loss and R^2 Score of the model on the DataLoader
def model_loss(model, dataset, train = False, optimizer = None):
  #It cycles through the batches and get the average L1 Loss
  performance = L1Loss()
  score_metric = R2Score()
  avg_loss = 0
  avg_score = 0
  count = 0
  for input, output in iter(dataset):
    #It gets the model's predictions 
    predictions = model.feed(input)
    #It gets the model's loss
    loss = performance(predictions, output)
    #It gets the model's R^2 score
    score_metric.update([predictions, output])
    score = score_metric.compute()
    if(train):
      #It clears the errors so that they won't cummulate
      optimizer.zero_grad()
      #It computes the gradients for optimizer
      loss.backward()
      #It uses the optimizer for updating the model's parameters based on the gradients
      optimizer.step()
    #It stores the loss and update the counter  
    avg_loss += loss.item()
    avg_score +=score
    count += 1
  return avg_loss / count, avg_score / count
```

```
#It defines the total number of epochs that are to be trained   
epochs = 200
#It defines the performance measure and the optimizer
optimizer = torch.optim.RMSprop(model.parameters(), lr=1e-2)
#It converts the training set into the torch variables to our model uisng the GPU as floats
inputs = torch.from_numpy(x_train_np).cuda().float()
#The reshape is used in removing a warning PyTorch outputs 
outputs = torch.from_numpy(y_train_np.reshape(y_train_np.shape[0], 1)).cuda().float()
#It creates the DataLoader to work
tensor = TensorDataset(inputs, outputs)
loader = DataLoader(tensor, batch_size, shuffle=True, drop_last=True)
#Starting the training loop
for epoch in range(epochs):
  #It cycles through the batches and gets the average loss 
  avg_loss, avg_r2_score = model_loss(model, loader, train=True, optimizer=optimizer)
  #Prints the average loss
  print("Epoch " + str(epoch + 1) + ":\n\tLoss = " + str(avg_loss) + "\n\tR^2Score = " + str(avg_r2_score))
```

```
#It converts the testing set into torch variables for our model by using the GPU as float
inputs = torch.from_numpy(x_test_np).cuda().float()
outputs = torch.from_numpy(y_test_np.reshape(y_test_np.shape[0], 1)).cuda().float()
#It creates the DataLoader instance to work with the batches
tensor = TensorDataset(inputs, outputs)
loader = DataLoader(tensor, batch_size, shuffle=True, drop_last=True)
#It prints the average performance i.e. both L1Loss and R^2 Score
avg_loss, avg_r2_score = model_loss(model, loader)
print("The model's L1 loss is: " + str(avg_loss))
print("The model's R^2 score is: " + str(avg_r2_score))
```
